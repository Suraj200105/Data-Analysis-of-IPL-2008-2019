```python
r=pd.DataFrame([],columns=['Team','Decision','Times']) 
r
```

```python
t= match['toss_winner'].unique()
t
```

```python
for i in t:
    temp_bat=match[(match['toss_winner']==i) & (match['toss_decision']=='field')]
    temp_ball=match[(match['toss_winner']==i) & (match['toss_decision']=='bat')]
    r=r.append({'Team':i,'Decision':'bat','Times':temp_bat['winner'].count()},ignore_index=True)
    r=r.append({'Team':i,'Decision':'ball','Times':temp_ball['winner'].count()},ignore_index=True)
```

```python
r
```

```python
sns.catplot(x='Team',y='Times',hue='Decision',data=r,kind='bar',height=5,aspect=2)
plt.xticks(rotation=90)
plt.xlabel('Teams',fontsize=15)
plt.ylabel('Total Toss Won',fontsize=15)
plt.title('Toss Decision Chart',fontsize=20)
plt.show()
```

```python
delivery= pd.read_csv('archive2\deliveries.csv')
```

```python
total_runs=delivery.groupby(['match_id','inning','batting_team','bowling_team']).sum().reset_index()
total_runs.drop(['over','ball','is_super_over','wide_runs','bye_runs','legbye_runs','noball_runs','penalty_runs'],axis=1,inplace=True)
```

```python
total_runs
```

```python
t=pd.merge(match,total_runs,how='left',left_on='id',right_on='match_id')
t.drop(['match_id'],axis=1,inplace=True)
t
```

```python
first_innings=t[t['inning']==1]
city1=first_innings.groupby(['city']).sum()['total_runs'].reset_index()
city1.sort_values(by='total_runs',ascending=False)
city1
```

```python
second_innings=t[t['inning']==2]
#second_innings.to_csv('2nd_innnings.csv')
city2=second_innings.groupby(['city']).sum()['total_runs'].reset_index()
city2.sort_values(by='total_runs',ascending=False)
```

```python
plt.figure(figsize=(13,6))
plt.plot(city2['city'],city2['total_runs'],marker='o')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
city=pd.merge(city1,city2,on='city')
city['sum']=city['total_runs_x']+city['total_runs_y']
city=city.sort_values(by='sum',ascending=True)
city
```

```python
x = np.arange(len(city['city']))
width = 0.3 
plt.figure(figsize=(20,15))
f=plt.bar(x - width/2, city['total_runs_x'], width, label='first_innings')
s=plt.bar(x + width/2, city['total_runs_y'], width, label='second_innings')
plt.ylabel('Scores',fontsize=20)
plt.xlabel('Cities',fontsize=20)
plt.title('Innings Scores by Cities',fontsize=30)
plt.xticks(x,city['city'])
plt.legend(fontsize=15)
plt.xticks(rotation=90)
plt.bar_label(f,padding=8,rotation=90)
plt.bar_label(s,padding=8,rotation=90)
plt.show()
```
